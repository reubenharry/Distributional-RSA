```python
import pickle
import pandas as pnd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
from dist_rsa.models.debugging.l1_discrete import l1_model
```

```python
def titrate(size,amount,l1_sig1):
    world_samples = l1_model(subj="subj1",pred="pred2",sig1=0.1,sig2=0.1,l1_sig1=l1_sig1,resolution=(size,amount))
    hm = pnd.DataFrame(world_samples,columns=np.arange(-size,size)+amount,index=np.arange(-size,size)*amount)
    seaborn.heatmap(hm)
```

```python
titrate(size=100,amount=0.1,l1_sig1=1.0)
```
