```python
import pickle
import pandas as pnd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
from dist_rsa.models.debugging.l1_discrete import l1_model
%matplotlib inline
```

```python
def titrate(size,amount,l1_sig1,quds,only_trivial,pred):
    world_samples = l1_model(subj="subj1",
                             pred=pred,
                             sig1=0.1,
                             sig2=0.1,
                             l1_sig1=l1_sig1,
                             resolution=(size,amount),
                             quds=quds,
                             only_trivial=only_trivial
                            )
    hm = pnd.DataFrame(world_samples,columns=np.arange(-size,size)*amount,index=np.arange(-size,size)*amount)
    seaborn.heatmap(hm)
```

```python
titrate(size=100,amount=0.1,l1_sig1=100.0,quds=['qud1'],pred='pred2',only_trivial=False)
```

```python
titrate(size=100,amount=0.1,l1_sig1=100.0,quds=['qud2'],pred='pred1',only_trivial=False)
```
