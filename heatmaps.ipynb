```python
import pickle
import pandas as pnd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
from dist_rsa.models.debugging.l1_discrete import l1_model
%matplotlib inline
```

```python
def titrate(size,amount,l1_sig1,quds,only_trivial,pred,just_s1,sig1,sig2):
    world_samples = l1_model(subj="subj1",
                             pred=pred,
                             sig1=sig1,
                             sig2=sig2,
                             l1_sig1=l1_sig1,
                             resolution=(size,amount),
                             quds=quds,
                             only_trivial=only_trivial,
                             just_s1=just_s1
                            )
    hm = pnd.DataFrame(world_samples,columns=np.arange(-size,size)*amount,index=np.arange(-size,size)*amount)
    seaborn.heatmap(hm)
    
```
