```python
import dill as pickle
metaphor_to_score_count = pickle.load(open("metaphor_to_score_count.pkl",'rb'))
```

```python
import numpy as np
import matplotlib.pyplot as plt
# width of the bars

def compare_l1_baseline(met):
    return metaphor_to_score_count["l1"][met][0]-metaphor_to_score_count["baseline"][met][0]

X_labels = sorted(list(metaphor_to_score_count["l1"]),key=lambda met: compare_l1_baseline(met),reverse=True)[:50]


# print(X_labels)
# print(compare_l1_baseline(X_labels[0]))

# raise Exception

barWidth = 0.1
 
# Choose the height of the blue bars
bars1 = [metaphor_to_score_count["l1"][met][0] for met in X_labels]
 
# Choose the height of the cyan bars
bars2 = [metaphor_to_score_count["baseline"][met][0] for met in X_labels]

width=0.5
# The x position of bars
r1 = width*np.arange(len(bars1))
r2 = [x + barWidth for x in r1]

plt.figure(figsize=(20,10))
# Create blue bars
plt.bar(r1, bars1, width = barWidth, color = 'blue', edgecolor = 'black', capsize=7, label='l1')
 
# Create cyan bars
plt.bar(r2, bars2, width = barWidth, color = 'cyan', edgecolor = 'black', capsize=7, label='baseline')
 
# general layout
plt.xticks([width*r + barWidth for r in range(len(bars1))], X_labels,rotation="vertical")
plt.ylabel('height')
plt.legend()

```

```python
X_labels = sorted(list(metaphor_to_score_count["l1"]),key=lambda met: compare_l1_baseline(met),reverse=True)[50:]


# print(X_labels)
# print(compare_l1_baseline(X_labels[0]))

# raise Exception

barWidth = 0.1
 
# Choose the height of the blue bars
bars1 = [metaphor_to_score_count["l1"][met][0] for met in X_labels]
 
# Choose the height of the cyan bars
bars2 = [metaphor_to_score_count["baseline"][met][0] for met in X_labels]

width=0.5
# The x position of bars
r1 = width*np.arange(len(bars1))
r2 = [x + barWidth for x in r1]

plt.figure(figsize=(20,10))
# Create blue bars
plt.bar(r1, bars1, width = barWidth, color = 'blue', edgecolor = 'black', capsize=7, label='l1')
 
# Create cyan bars
plt.bar(r2, bars2, width = barWidth, color = 'cyan', edgecolor = 'black', capsize=7, label='baseline')
 
# general layout
plt.xticks([width*r + barWidth for r in range(len(bars1))], X_labels,rotation="vertical")
plt.ylabel('height')
plt.legend()
```

```python
# X_labels = sorted(list(metaphor_to_score_count["l1"]),key=lambda met: compare_l1_baseline(met),reverse=True)[50:]


# print(X_labels)
# print(compare_l1_baseline(X_labels[0]))

# raise Exception

mean_difference = {}

for met in list(metaphor_to_score_count["l1"]):
    l1_scores = metaphor_to_score_count["l1"][met][0]
    l1_count = metaphor_to_score_count["l1"][met][1]
    baseline_scores = metaphor_to_score_count["baseline"][met][0]
    baseline_count = metaphor_to_score_count["baseline"][met][1]
    mean_difference[met] = (l1_scores/l1_count) - (baseline_scores/baseline_count)

X_labels = sorted(list(mean_difference),key=lambda x: mean_difference[x])

barWidth = 0.1
 
# Choose the height of the blue bars
bars1 = [mean_difference[met] for met in X_labels]
 
# Choose the height of the cyan bars
# bars2 = [metaphor_to_score_count["baseline"][met][0] for met in X_labels]

width=0.3
# The x position of bars
r1 = width*np.arange(len(bars1))
# r2 = [x + barWidth for x in r1]

plt.figure(figsize=(20,7))
# Create blue bars
plt.bar(r1, bars1, width = barWidth, color = 'blue', edgecolor = 'black', capsize=15)
 
# Create cyan bars
# plt.bar(r2, bars2, width = barWidth, color = 'cyan', edgecolor = 'black', capsize=7, label='baseline')
 
# general layout
plt.xticks([width*r + barWidth for r in range(len(bars1))], X_labels,rotation="vertical")
plt.ylabel('Difference between Mean of L1Q and Baseline Scores')
plt.legend()
```
