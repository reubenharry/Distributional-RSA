```python
import pickle
import pandas as pnd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
from dist_rsa.models.debugging.l1_discrete import l1_model
%matplotlib inline
```

```python
def titrate(size,amount,l1_sig1,quds,only_trivial,pred,just_s1,just_l0=False,sig1=0.1,sig2=0.1,prior=False):
    results = l1_model(subj="subj1",
                             pred=pred,
                             sig1=sig1,
                             sig2=sig2,
                             l1_sig1=l1_sig1,
                             resolution=(size,amount),
                             quds=quds,
                             only_trivial=only_trivial,
                             just_s1=just_s1,
                             just_l0=just_l0
                            )


#     n=np.asarray([1 if x[1]==3 or x[0]==3 else 0 for x in results[1]])
# #     print(results[0].shape)
# #     print(n.shape)
#     print("\n\n\nRESULTS\n\n\n")
#     print(n)
#     n=np.reshape(n,(size*2,size*2))
    
#     hm = pnd.DataFrame(n,columns=np.arange(-size,size)*amount,index=np.arange(-size,size)*amount)
#     seaborn.heatmap(hm)
    
#     return None

    if not prior:
        results[0][size,:]=1
        results[0][:,size]=1
        results[0][size-1,size]=2
        results[0][size,size+1]=2
        hm = pnd.DataFrame(results[0],columns=np.arange(-size,size)*amount,index=np.arange(-size,size)*amount)
        seaborn.heatmap(hm)
    
#     hm = pnd.DataFrame(results[1],columns=np.arange(-size,size)*amount,index=np.arange(-size,size)*amount)
#     seaborn.heatmap(hm)

    else:

        hm = pnd.DataFrame(results[2],columns=np.arange(-size,size)*amount,index=np.arange(-size,size)*amount)
        seaborn.heatmap(hm)
    
```

```python
titrate(size=100,amount=0.1,l1_sig1=1.0,quds=['qud1'],pred='pred2',only_trivial=False,just_s1=True,just_l0=False,sig1=2.0,sig2=2.0,prior=False)
```
