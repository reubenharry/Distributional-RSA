```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
 
# Create data
def show_plot(a,min_scale,max_scale,name):
    x = a[:,0]
    y = a[:,1]


    # Plot
    plt.scatter(x, y)
    plt.title(name)
    plt.xlabel('x')
    plt.ylabel('y')
    
    plt.xlim(min_scale,max_scale)
    plt.ylim(min_scale,max_scale)
    
    plt.show()


for i in range(20):
    a = pickle.load(open("dist_rsa/models/debugging/l1_sig1-0.1/world_samples_subj1pred1"+str(i),'rb'))
    b = pickle.load(open("dist_rsa/models/debugging/l1_sig1-10/world_samples_subj1pred1"+str(i),'rb'))
    c = pickle.load(open("dist_rsa/models/debugging/l1_sig1-0.1/world_samples_subj2pred1"+str(i),'rb'))
    d = pickle.load(open("dist_rsa/models/debugging/l1_sig1-10/world_samples_subj2pred1"+str(i),'rb'))
    show_plot(a,-2,2,name="l1_sig:0.1, subj1pred1, iteration:"+str(i))
    show_plot(b,-20,20,name="l1_sig:10.0, subj1pred1, iteration:"+str(i))
    show_plot(c,-2,2,name="l1_sig:0.1, subj2pred1, iteration:"+str(i))
    show_plot(d,-20,20,name="l1_sig:10.0, subj2pred1, iteration:"+str(i))
    
```
